4. Python Program for factorial of a number

```python
def fact(n):
    fact1 = 1
    for j in range(1, n+1):
        fact1 = fact1 * j
    return fact1
x = int(input("Enter the num:"))
out1 = fact(x)
print("The factorial of", x, "is", out1)
```

5. Python Program for simple interest

```python
def interest(p, r, t):
    SI = (p*r*t)/100
    return SI
principal_amount = float(input("Enter the proncipal amount:"))
ROI = float(input("Enter the percent rate of interest per annum:"))
interest_period = float(input("Enter the time in years"))
out1 = interest(principal_amount, ROI, interest_period)
print("The principal interest for amount =", principal_amount, 
      "with rate of interest =", ROI,"% and interest period =", 
      interest_period, "is =", out1)
```

6. Python Program for compound interest

```python
def interest(p, r, n, t):
    CI = (p*((1+(r/(100 * n)))**(n*t)))-p
    return CI
principal_amount = float(input("Enter the proncipal amount:"))
ROI = float(input("Enter the percent rate of interest per annum:"))
comp_freq = float(input("Enter the compounding frequency"))
interest_period = float(input("Enter the time in years"))
out1 = interest(principal_amount, ROI, comp_freq, interest_period)
print("The principal interest for amount =", principal_amount, 
      "with rate of interest =", ROI,"% compounding frequency =",
      comp_freq, "and interest period =", interest_period, "is =", out1)
```

8. Python Program for Program to find area of a circle

```python
import math
def Area(r):
    pi = math.pi
    area = pi * (r **2)
    return area
rad = float(input("Enter the radius of the circle ="))
result = Area(rad)
print("The area of a circle of radius", rad, "is =", result)
```

9. Python program to print all Prime numbers in an Interval

```python
lower_ran = int(input("Enter the lower range:"))
upper_ran = int(input("Enter the upper range:"))
print("In the given range between lower range", lower_ran, "and upper range",
     upper_ran, "the prime numbers are as folows:")
for num in range(lower_ran, upper_ran):
    if num > 1:
        for i in range(2, num):
            if num % i == 0 :
                break
        else:
            print(num)
```

10. Python program to check whether a number is Prime or not

```python
n = int(input("Enter the given number:"))
count = 0
if n > 1:
    for i in range(2,n):
        if(n % i == 0):
            count += 1
            break
    else:       
        print("n =", n, "is a prime number")
    if(count != 0):            
        print("n =", n, "is not a prime number")            
```
